 1. What is the difference between interpreted and compiled languages?
  >Compiled Languages:  
* The entire program is translated before it runs.  
* Because this translation happens ahead of time, compiled programs tend to run faster and use less memory at runtime.   
* Examples: C, C++, Rust, Go.

  >Interpreted Languages:
*  Code is translated and executed at runtime (line by line or statement by statement).
*  This makes development faster and more flexible (you can often run code immediately without compiling), but it usually runs slower because translation happens as the program runs .
*   Examples: Python, JavaScript, Ruby.

18.  What is the difference between os.fork() and multiprocessing in Python ?

> os.fork() :
* Creates a child process by duplicating the current process.
* Available only on Unix/Linux systems; not supported on Windows.
* Requires manual handling of inter-process communication (IPC).
* Lower-level, closer to the OS.
* Pros: Lightweight, simple for Unix-specific tasks
* Cons: Not portable, manual management of IPC

> multiprocessing:
* A high-level API to create processes in a cross-platform way.
* Works on Windows, Linux, macOS.
* Handles process creation, IPC, queues, pools automatically.
* Pros: Portable, easy IPC, multiple features
* Cons: Slightly more overhead than os.fork()

```python
#(6)Write a Python program that logs an error message to a log file when a division by zero exception occur?
import logging

for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    filename="error.log",
    level=logging.ERROR,
    format="%(levelname)s - %(message)s"
)

try:
    a = 0
    result = 10 / a
    print("Result:", result)

except ZeroDivisionError:
    logging.error("Division by zero error occurred")
    print("Error: Division by zero is not allowed")

with open("error.log", "r") as file:
    print("\nLog file content:")
    print(file.read())

```

```python
#(18)How would you implement a basic logging setup that logs to a file with rotation after 1MB?
import logging
from logging.handlers import RotatingFileHandler

for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

handler = RotatingFileHandler(
    filename="app.log",
    maxBytes=1 * 1024 * 1024,
    backupCount=3
)

formatter = logging.Formatter(
    "%(asctime)s - %(levelname)s - %(message)s"
)
handler.setFormatter(formatter)

logger = logging.getLogger()
logger.setLevel(logging.INFO)
logger.addHandler(handler)

for i in range(10000):
    logger.info(f"Logging message number {i}")

print("Logging completed. Check app.log and rotated files.")


```

```python
#(21)Write a Python program that reads a file and prints the number of occurrences of a specific word ?
with open("sample.txt", "w") as file:
    file.write("Python is great.\n")
    file.write("I love Python programming.\n")
    file.write("Python is easy to learn.\n")

word_to_count = "Python"
count = 0

with open("sample.txt", "r") as file:
    for line in file:
        words = line.split()
        count += words.count(word_to_count)
print(f"The word '{word_to_count}' occurs {count} times in the file.")
```

```python
#(22) How can you check if a file is empty before attempting to read its contents?
import os

file_name = "sample.txt"

with open(file_name, "w") as file:
    pass

if os.path.exists(file_name) and os.path.getsize(file_name) > 0:
    with open(file_name, "r") as file:
        content = file.read()
        print("File content:")
        print(content)
else:
    print("The file is empty.")

```

```python
#(23)Write a Python program that writes to a log file when an error occurs during file handling?
import logging

for handler in logging.root.handlers[:]:
    logging.root.removeHandler(handler)

logging.basicConfig(
    filename="file_error.log",
    level=logging.ERROR,
    format="%(asctime)s - %(levelname)s - %(message)s"
)

try:
    with open("non_existent_file.txt", "r") as file:
        content = file.read()
        print(content)

except FileNotFoundError:
    logging.error("FileNotFoundError: Attempted to open a file that does not exist.")
    print("Error: File does not exist.")

except Exception as e:
    logging.error(f"Unexpected error: {e}")
    print("An unexpected error occurred.")

finally:
    print("File handling attempt completed.")

with open("file_error.log", "r") as log_file:
    print("\nLog file content:")
    print(log_file.read())



```
